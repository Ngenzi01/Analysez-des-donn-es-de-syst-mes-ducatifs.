### II) Description des informations dans le jeu de données.  

```python
def get_data_description(data3):
    """ Compte le nombre de ranges dans le fichier EdStatsData.csv"""
    #print(data3.axes)             ##une liste représentant les axes du dataframe.
    nb_ranges = len(data3.axes[0])       #Compte le nombre de rangées
    nb_cols = len(data3.axes[1])         #Compte le nombre de colonnes 
    print("Nombre de rangées: ", nb_ranges)
    print("Number of Columns: ", nb_cols)
```

### III) Selection des informations pertinentes

####  commencons par l'elimination des valeurs manquantes

```python
def clean_data(data3):
    """ cette fonction permet d'effacer les valeurs manquantes"""
    data3DeleteNANColumns = data3.drop(data3.loc[:, '2015':'2100'].columns, axis=1)
    data3DeleteNANRows=data3DeleteNANColumns.dropna(thresh=30)# deleting columns with more than 30 values missing 
    data3DeleteNANColumnsNew=data3DeleteNANRows.dropna(axis=1,how='all') # drop columns with NAN values on every row
    data3Cleaned=data3DeleteNANColumnsNew.dropna().reset_index(drop=True) # Reset index after drop
    print(" Information sur le fichier cleané")
    print("                                    ")
    #data3Cleaned.info()  #verfier si il n y aplus de valeur manquante
    print("Maintenant le fichier cleané contient :")
    print("-------------------------------------")
    get_data_description(data3Cleaned)
    #data3Cleaned.to_csv("/Users/ngenzi/Desktop/Project_2_openclassroom/Projet+Python_Dataset_Edstats_csv/CleanedData.csv")
    return data3Cleaned
```

### Analysons les indiateurs  statistique par region

```python
def select_zone(Row1,Row2,Cleaneddataframe):
    """Cette fonction permet de sélectionner la zone à étudier"""
    zone=Cleaneddataframe.iloc[Row1:Row2] #j'explique ce que fait précisément cette ligne 
    zonereindexed =zone.reset_index()
    return zonereindexed
```

### Conversion des données dans un tableau numpy

```python
def numpy_conversion(Row,Col,data):
    """cette fincrtion perment de convertir les rangées selectionnées dans un tableau numpy"""
    SelectedRow=data.iloc[Row,Col:]
    DataArray= SelectedRow.to_numpy()
    return DataArray
```

```python
def axis_selection(StartYear,EndYear,DataArray):
    """ cette fonction perment de selection les donnees """
    year= np.linspace(StartYear,EndYear,len(DataArray))
    return year
```

#### Tracer les courbes

```python
def plot_data(datax,datay):
    """ cette fonction permet de tracer les courbes"""
    plt.plot(datax,datay,linewidth=2)
    plt.xlabel('Year')
    plt.xlim([1970,2015])
```

####  Tracer les pie digrams

```python
def plot_barchart(datax,datay):
    """ c'est une fonction qui permet de plotter les donnée en barchart """
    plt.bar(datax, datay)
```

#### Tracer les box plots

```python
def figure_custom():
    """figure customisation"""
    fig = plt.figure(figsize =(12, 7))
    ax = fig.add_subplot(111)
    return fig,ax
```

```python
def Box_plot(data,fig,ax):
    """Cette fonction permet de tracer les diagrame de moustaches"""
    
    # Creating axes instance
    bp = ax.boxplot(data, patch_artist = True,notch ='True', vert = 0)
    colors = ['#0000FF', '#00FF00','#FFFF00', '#FF00FF']
    
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
    # changing color and linewidth of
    # whiskers
    
    for whisker in bp['whiskers']:
        whisker.set(color ='#8B008B',linewidth = 1.5,linestyle =":")
        # changing color and linewidth of
        # caps
    for cap in bp['caps']:
        cap.set(color ='#8B008B',linewidth = 2)
        # changing color and linewidth of
        
    # medians
    
    for median in bp['medians']:
        median.set(color ='red',linewidth = 3)
    
    # changing style of fliers
    
    for flier in bp['fliers']:
        flier.set(marker ='D',color ='#e7298a',alpha = 0.5)
    # Removing top axes and right axes
    # ticks
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

#### Plotting Piechart diagram

```python
def plot_pieChart(data1,data2):
    """Cette fonction permet de tracer les donne en piechart"""
    fig = plt.figure(figsize =(10, 7))
    plt.pie(data1, labels = data2)
    plt.show()
```
